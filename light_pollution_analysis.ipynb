```python
import plotly.express as px

# Define a function to calculate growth between the first and last years with percentages
def calculate_growth_with_percent(df, column):
    earliest_year = df['Year'].min()
    latest_year = df['Year'].max()

    start_values = df[df['Year'] == earliest_year].groupby('NAME')[column].mean()
    end_values = df[df['Year'] == latest_year].groupby('NAME')[column].mean()

    growth = end_values - start_values
    percent_growth = (growth / start_values.replace(0, pd.NA)) * 100

    growth_df = pd.DataFrame({
        'Absolute Growth': growth,
        'Percent Growth': percent_growth
    })

    return growth_df

# Calculate growth for population and light pollution
population_growth = calculate_growth_with_percent(df, 'Population')
light_pollution_growth = calculate_growth_with_percent(df, 'light_pollution')

# Merge for comparison
growth_comparison = population_growth.join(light_pollution_growth, lsuffix='_population', rsuffix='_light_pollution').dropna().reset_index()

# Filter for range: population growth between 5k and 100k, and light pollution > 0
growth_comparison_filtered = growth_comparison[
    (growth_comparison['Absolute Growth_population'] >= 5000) &
    (growth_comparison['Absolute Growth_population'] <= 100000) &
    (growth_comparison['Absolute Growth_light_pollution'] > 0)
]

# Plotting with Plotly using log scale for clarity
fig = px.scatter(
    growth_comparison_filtered,
    x='Absolute Growth_population',
    y='Absolute Growth_light_pollution',
    text='NAME',
    hover_name='NAME',
    size='Absolute Growth_population',
    color='Absolute Growth_light_pollution',
    trendline='ols',
    log_x=True,
    title='Population Growth vs. Light Pollution Growth (Logarithmic Scale)',
    labels={
        'Absolute Growth_population': 'Population Growth (5K–100K, Log Scale)',
        'Absolute Growth_light_pollution': 'Light Pollution Growth (cd·m⁻²)'
    },
    width=1400,
    height=800,
    color_continuous_scale='Plasma'
)

fig.update_traces(textposition='top center', textfont=dict(size=10))
fig.update_layout(
    xaxis_title='Population Growth (5K–100K, Log Scale)',
    yaxis_title='Light Pollution Growth (cd·m⁻²)',
    coloraxis_colorbar_title='Light Pollution Growth (cd·m⁻²)',
    template='plotly_white',
    margin=dict(t=50, b=50, l=50, r=50)
)

fig.show()
```

Points below the trendline indicate towns where:

Light pollution grew less than expected for their level of population growth.

So if a city grew a lot in population but didn’t see a proportionally high increase in light pollution, it falls below the line.

This can be due to:

Energy-efficient lighting upgrades (e.g., LED retrofits)

Effective zoning and light pollution ordinances

Different urban development patterns (e.g., less sprawl, fewer commercial zones)

# Yearly (Time-Series) Trends:
1.	Time-Series of Light Pollution
o	Showcase how total light pollution (ntl_sum) evolved over time for your top 5 fastest-growing towns.


```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

# Load data


# Extract year if needed
if 'Year' not in df.columns:
    df['Year'] = pd.to_datetime(df['date']).dt.year

# Function to calculate growth
def calculate_growth_with_percent(df, column):
    start = df['Year'].min()
    end = df['Year'].max()
    start_vals = df[df['Year'] == start].groupby('NAME')[column].mean()
    end_vals = df[df['Year'] == end].groupby('NAME')[column].mean()
    growth = end_vals - start_vals
    percent = (growth / start_vals.replace(0, pd.NA)) * 100
    return pd.DataFrame({'Absolute Growth': growth, 'Percent Growth': percent})

# Top 10 towns by population growth
growth_df = calculate_growth_with_percent(df, 'Population')
top10_towns = growth_df['Absolute Growth'].sort_values(ascending=False).head(10).index.tolist()

# Filter and group for plotting
top10_df = df[df['NAME'].isin(top10_towns)]
grouped = top10_df.groupby(['NAME', 'Year'], as_index=False)['ntl_sum'].mean()

# Use plasma color palette
plasma_colors = px.colors.sequential.Plasma
color_map = {name: plasma_colors[i % len(plasma_colors)] for i, name in enumerate(top10_towns)}

# Create figure
fig = go.Figure()

for town in top10_towns:
    town_data = grouped[grouped['NAME'] == town]
    fig.add_trace(go.Scatter(
        x=town_data['Year'],
        y=town_data['ntl_sum'],
        mode='lines+markers',
        name=town,
        line=dict(color=color_map[town], width=2),
        marker=dict(size=6)
    ))

# Update layout
fig.update_layout(
    yaxis_type='log',
    title='Yearly Trend of Average Light Pollution (NTL Sum, Log Scale) in Top 10 Fastest-Growing Towns',
    xaxis_title='Year',
    yaxis_title='Average NTL Sum (nW·cm⁻²·sr⁻¹, log scale)',
    template='plotly_white',
    width=1400,
    height=800,
    legend_title_text='Town'
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load your dataset


# Ensure Year column exists
if 'Year' not in df.columns:
    df['Year'] = pd.to_datetime(df['date']).dt.year

# Calculate light pollution per capita
df['light_pollution_per_capita'] = df['light_pollution'] / df['Population']

# Functions to calculate raw growth and percent growth
def get_growth(df, col):
    start = df[df['Year'] == df['Year'].min()].groupby('NAME')[col].mean()
    end = df[df['Year'] == df['Year'].max()].groupby('NAME')[col].mean()
    return (end - start).dropna()

def get_percent_growth(df, col):
    start = df[df['Year'] == df['Year'].min()].groupby('NAME')[col].mean()
    end = df[df['Year'] == df['Year'].max()].groupby('NAME')[col].mean()
    return ((end - start) / start).dropna() * 100

# Get raw population growth for top 10 selection
pop_growth = get_growth(df, 'Population')

# Calculate percent growth
pop_pct_growth = get_percent_growth(df, 'Population')
ntl_pct_growth = get_percent_growth(df, 'ntl_sum')
lp_per_capita_pct_growth = get_percent_growth(df, 'light_pollution_per_capita')

# Combine into a DataFrame
percent_growth_df = pd.DataFrame({
    'Population Growth (%)': pop_pct_growth,
    'Light Pollution Growth (%)': ntl_pct_growth,
    'Per Capita LP Growth (%)': lp_per_capita_pct_growth
}).dropna()

# Select top 10 towns by raw population growth
top10_names = pop_growth.sort_values(ascending=False).head(10).index
top10_percent = percent_growth_df.loc[top10_names]

# Melt for plotting
percent_long = top10_percent.reset_index().melt(id_vars='NAME', var_name='Metric', value_name='Percent Growth')

# Colors
colors = {
    'Population Growth (%)': '#1f77b4',
    'Light Pollution Growth (%)': '#9467bd',
    'Per Capita LP Growth (%)': '#e377c2'
}

# Plot
fig = go.Figure()

for metric in percent_long['Metric'].unique():
    metric_data = percent_long[percent_long['Metric'] == metric]
    fig.add_trace(go.Bar(
        x=metric_data['Percent Growth'],
        y=metric_data['NAME'],
        orientation='h',
        name=metric,
        marker_color=colors[metric],
        text=metric_data['Percent Growth'].round(1).astype(str) + '%',
        textposition='auto'
    ))

fig.update_layout(
    title='Percent Growth Comparison (Top 10 Towns by Population Growth)',
    xaxis_title='Percent Growth (%)',
    yaxis_title='Town',
    barmode='group',
    height=800,
    width=1000,
    template='plotly_white'
)

fig.show()
```
